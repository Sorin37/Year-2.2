```python
n = 100 ; x2 = np.array(range(1, n+1))
x1 = .01 * x2 + np.random.uniform(-.1, .1, size=n)
y = -x1 + x2 + np.random.randn(n)*.01

df = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
lm1 = smf.ols(formula='y ~ x1', data=df).fit()
lm2 = smf.ols(formula='y ~ x1 + x2', data=df).fit()
print(lm1.summary())
print(lm2.summary())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(x2, x1, 'o')
ax.set_xlabel('x2')
ax.set_ylabel('x1')
plt.show()
```

```python
newdf = pd.DataFrame({'y': y, 'x1': x1, 'x2': x2,
    'ey': smf.ols(formula='y ~ x2', data=df).fit().resid,
    'ex1': smf.ols(formula='x1 ~ x2', data=df).fit().resid})
params = smf.ols(formula='y ~ x1', data=df).fit().params
x = np.linspace(0, 1, 10)
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
cm = plt.cm.get_cmap('winter')
sc = ax.scatter(x=x1, y=y, c=x2, cmap=cm)
ax.plot(x, params[0] + x * params[1])
ax.set_xlabel('x1') ; ax.set_ylabel('y')
fig.colorbar(sc, ax=ax).set_label('x2')
ax.grid() ; plt.show()
```

```python
params = smf.ols(formula='ey ~ ex1', data=newdf).fit().params
x = np.linspace(-0.1, 0.1, 10)
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
cm = plt.cm.get_cmap('winter')
sc = ax.scatter(x=newdf.ex1.values, y=newdf.ey.values, c=newdf.x2.values, cmap=cm)
ax.plot(x, params[0] + x * params[1])
ax.set_xlabel('ex1') ; ax.set_ylabel('ey')
fig.colorbar(sc, ax=ax).set_label('x2')
plt.show()
```

__Q1__. Pentru swiss dataset, realizai modelul de regresie liniar pentru Fertility n funcie de Agriculture i Education. Remarcai intercept i cei doi coeficieni.

Mai departe artai cum coeficientul regresorului $x_1$ (Agriculture) este de fapt coeficientul regresiei caracteristice rspunsului $y$ (Fertility) i regresorului $x_2$ (Education) din care contribuia celuilalt regresor ($x_1$) a fost deja nlturat.

__Hint1__: contribuia unui regresor nu mai este prezent n reziduul asociat, i se nltur folosind regresia liniar (revedei slide-urile 42-end din cursul 7).

__Hint2__: folosii proprietatea .resid a modelului deja potrivit.

```python
swiss = pd.read_csv('swiss.csv')
swiss.columns = ['District', 'Fertility', 'Agriculture', 'Examination', 'Education', 'Catholic', 
                 'InfantMortality']
swiss.head()
```

```python
_lm = smf.ols(formula='Fertility ~ Agriculture + Education', data = swiss).fit()
_lm.summary()
```

```python
lm1 = smf.ols(formula='Fertility ~ Education', data = swiss).fit()
lm2 = smf.ols(formula='Agriculture ~ Education', data = swiss).fit()

swiss['residFertility'] = lm1.resid
swiss['residAgriculture'] = lm2.resid
lm = smf.ols(formula='residFertility ~ residAgriculture -1', data = swiss).fit()
lm.summary()
```

```python
print('PUNCTUL A:')
nr_cilindri = mtcars.cyl
sns.pairplot(mtcars);
```

```python
print('PUNCTUL B:')
lm = smf.ols(formula='mpg ~ C(cyl) + disp + hp + drat + wt + qsec + vs + am + gear + carb', data = mtcars).fit()
lm.summary()
```

```python
print('PUNCTUL C:')
y = mtcars['mpg'].values
x = mtcars['hp'].values
beta1 = np.corrcoef(x,y)[0,1] * np.std(y)/np.std(x)
beta0 = np.mean(y) - beta1 * np.mean(x)
y1 = beta1*x+beta0
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.scatter(x, y, c='k', alpha = .2, s=50)
ax.plot(x, y1, lw=4)
ax.set(xlabel="hp", ylabel="mpg")
ax.grid(True)
plt.show()
```

```python
print('PUNCTUL D:')
lm = smf.ols(formula='mpg ~ cyl + hp', data = mtcars).fit()
lm.summary()
```

```python
x = mtcars['hp'].values

params = smf.ols(formula='mpg ~ hp + cyl', data=mtcars).fit().params
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
c = mtcars['cyl'].copy()
c[mtcars['cyl'] == 4] = 'Red'
c[mtcars['cyl'] == 6] = 'Blue'
c[mtcars['cyl'] == 8] = 'Green'
ax.scatter(mtcars.hp, mtcars.mpg, c=c)
ax.plot(x, params[0] + params[1] * x + params[2] * 4, 'r')
ax.plot(x, params[0] + params[1] * x + params[2] * 6, 'b')
ax.plot(x, params[0] + params[1] * x + params[2] * 8, 'g')
ax.set_xlabel('hp') ; ax.set_ylabel('mpg')
ax.legend(['4cyl', '6cyl', '8cyl']) ; plt.grid() ; plt.show()
```

```python
smf.ols(formula='mpg ~ hp + cyl', data=mtcars).fit().summary()
```

```python
print('PUNCTUL E:')

x = mtcars['hp'].values
params = smf.ols(formula='mpg ~ hp * C(cyl)', data=mtcars).fit().params
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
c = mtcars['cyl'].copy()
c[mtcars['cyl'] == 4] = 'Red'
c[mtcars['cyl'] == 6] = 'Blue'
c[mtcars['cyl'] == 8] = 'Green'
ax.scatter(mtcars.hp, mtcars.mpg, c=c)
ax.plot(x, params[0] + params[1] * x + params[2] * 4 + params[3] * x * 4, 'r')
ax.plot(x, params[0] + params[1] * x + params[2] * 6 + params[3] * x * 6, 'b')
ax.plot(x, params[0] + params[1] * x + params[2] * 8 + params[3] * x * 8, 'g')
ax.set_xlabel('hp') ; ax.set_ylabel('mpg')
ax.legend(['4cyl', '6cyl', '8cyl']) ; plt.grid() ; plt.show()
```
