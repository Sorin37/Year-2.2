## Ridicarea distribuiei mediei aruncrilor cu zarul, prin numrare 

Ne propunem s determinm exact de cte ori poate iei o anumit sum a feelor din totalul sumelor posibile. De exemplu, dac aruncm de dou ori cu zarul, numrm urmtoarele:
* suma 2 poate iei pentru o singur combinaie, 1+1
* suma 3 poate iei pentru 2 combinaii, 1+2 i 2+1
* suma 4 poate iei pentru 3 combinaii, 1+3, 2+2 i 3+1
* suma 5 - 4 combinaii, 1+4, 2+3, 3+2, 4+1
* suma 6 - 5 combinaii, 1+5, 2+4, 3+3, 4+2, 5+1
* suma 7 - 6 combinaii, 1+6, 2+5, 3+4, 4+3, 5+2, 6+1 
* ...

n total sunt 1+2+3+4+5+6+5+4+3+2+1 = 36 sume posibile, dar numai 11 sunt distincte (de la 2 la 12). Distribuia lor ar arta aa:

```python
sume = np.array(range(2, 13)) 
count = np.array([1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1])
plt.bar(sume, count)
plt.xlabel('sume posibile')
plt.ylabel('de cte ori pot aprea\nn totalul celor %d variante' % np.sum(count))
plt.show()
```

Vom generaliza aceast procedur pentru un numr de $n=10$ aruncri cu zarul. Mai departe vom ridica histograma variabilei sum.

```python
# definim un zar ca un dictionar - fiecare fata (side) a zarului poate iesi o data
d = {}
for i in range(6):
    d[i+1] = 1
    
def cumulate(a, d):
    """
    Concatenam la posibilele sume din dictionarul a inca o aruncare;
    vom exprima toate combinatiile de sume care pot aparea, si vom numara
    numarul de aparitii a fiecarei sume
    """
    res = {}
    for k1, v1 in a.items():
        for k2, v2 in d.items():
            # aici ajungem sa numaram toate combinatiile posibile 
            # exprimate de cele doua dictionare
            res[k1+k2] = v1*v2 if res.get(k1+k2) is None else res[k1+k2] + v1*v2
    return res

# construim recursiv sumele posibile care pot aparea pentru n = 10 aruncari
# pentru doua aruncari, argumentul range() trebuie sa aiba valoarea 1
# puteti inlocui cu 1, trebuie sa obtinem aceeasi histograma de mai sus
t = d
for i in range(9):
    t = cumulate(t, d)

sums = np.array(list(t.keys()))
counts = np.array(list(t.values()))

plt.bar(sums, counts)
plt.xlabel('sume posibile')
plt.ylabel('de cte ori pot aprea\nn totalul celor %d variante' % np.sum(counts))
plt.show()
```

Pentru fiecare sum, calculm media counts-urilor; prin mprire la o constant, forma histogramei nu se schimb:

```python
means = counts / np.sum(counts)
plt.bar(sums, means)
plt.show()
```

Observai cum s-a schimbat doar scala (ordonata). Este aceast distribuie o densitate de probabilitate (PMF)? S calculm aria:

```python
np.sum(means)
```

Se pare c ceea ce am desenat este o probability mass function, chiar o distribuie normal!

Distribuia Poisson este o distribuie discret care exprim probabilitatea apariiei unui numr dat de evenimente ntr-un interval fix de timp (sau spaiu), dac aceste evenimente apar cu o rat medie cunoscut, independent de momentul de timp de la ultimul eveniment (Wikipedia).

PMF pentru distribuia Poisson are un singur parametru, $\lambda$. PMF calculeaz probabilitatea ca variabila aleatoare X, distribuit Poisson, s ia valoarea x. Anume:

$P(X=x) = \lambda^x e^{-\lambda} / x!$

x ia valori discrete de la 0 la infinit. Media i dispersia pentru variabila Poisson X sunt tot $\lambda$.

Variabilele Poisson sunt folosite pentru modelarea ratei defectelor, de exemplu rata de defectare a SSD-urilor. Scriem $X \sim Poisson(\lambda t)$, unde $\lambda$ este valoarea ateptat a count-ului pe unitatea de timp iar $t$ este timpul total de monitorizare.

De exemplu, presupunem c numrul de oameni care apar la o staie de autobuz este o variabil Poisson cu o medie de 2.5 pe or, i vrem s aflm probabilitatea ca cel mult 3 oameni s apar ntr-o perioad de 4 ore. Folosim funcia stats.poisson.cdf(.. , mu= ..) pentru a afla probabilitatea ca variabila aleatoare s fie mai mic sau egal cu 3. Indicm quantila (3) i media (2.5*4).

```python
stats.poisson.cdf(3, mu=2.5*4)
```

Q37. Pentru aruncarea repetat a unei monede, numrai, folosind exemplul histogramei aruncrilor zarului, de cte ori avem $k$ succese din 10 aruncri succesive.

Ce distribuie are suma feelor obinute?

Repetai acelai lucru cu extragerea cu reintroducere a unei cri din pachetul de cri de joc.

```python
d = {}
for i in range(2):
    d[i] = 1
    
def cumulate(a, d):
    res = {}
    for k1, v1 in a.items():
        for k2, v2 in d.items():
            res[k1+k2] = v1*v2 if res.get(k1+k2) is None else res[k1+k2] + v1*v2
    return res
t = d
for i in range(9):
    t = cumulate(t, d)

sums = np.array(list(t.keys()))
counts = np.array(list(t.values()))

plt.bar(sums, counts)
plt.xlabel('sume posibile')
plt.ylabel('de cte ori pot aprea\nn totalul celor %d variante' % np.sum(counts))
plt.show()
```
